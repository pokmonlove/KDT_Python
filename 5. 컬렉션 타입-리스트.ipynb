#3. 리스트의 메서드

리스트를 다루기 위한 여러 함수와 메소드들이 있습니다.

```python
print(len("파이썬")) ##
```

```python
#len(): 객체의 길이를 반환하는 파이썬의 기본 내장 함수
list6=[10,20,30]
print(len(list6))
```

```python
list6=[10,20,30]
print(list6)
#append(): 리스트 요소의 끝에 새로운 요소를 추가해 주는 메소드
list6.append(100)
print(list6)

#list6.append(200,300) #TypeError: list.append() takes exactly one argument (2 given)
#항상 두 개 이상의 데이터를 다룰 땐 리스트나 튜플을 이용해야 한다
list6.append([200,300])
print(list6) #귀찮더라도 하나씩 append 해야 한다
```

```python
list6=[10,20,30]
#extend(): 리스트의 요소의 끝에 새로운 여러 요소를 추가
#list6.extend(100) #TypeError: 'int' object is not iterable #순서가 있는 자료 타입으로 넣어 달라는 뜻
list6.extend([100])
print(list6)
list6.extend([100, 200, 300]) #여러 개의 데이터를 넣을 때 쓴다
print(list6)
```

```python
list6=[10,20,30]
print(list6)
#pop(): 리스트 마지막 요소를 삭제하고 삭제된 요소를 반환해 주는 함수
temp=list6.pop()
print(temp)
print(list6) #pop한 요소가 없어졌네
```

```python
list6=[10,20,30]
print(list6)
#index(): 리스트에서 특정 값의 인덱스를 반환
print(list6.index(30)) #30에 해당하는 인덱스 값을 찾아 준다
#print(list6.index(100)) #ValueError: 100 is not in list #값이 없으면 에러가 난다
```

```python
list7=[100,5,70,40,50]
print(list7)
#reverse(): 리스트의 요소들의 순서를 반대로 설정해 준다
list7.reverse()
print(list7)
```

```python
#슬라이싱을 사용하여 리스트의 요소들의 순서를 반대로 설정하는 방법
list7=[100,5,70,40,50]
print(list7[:])
print(list7[0:5])
print(list7[::])
print(list7[0:5:1]) #스텝을 생략 스텝의 기본값이 1이기 때문
print(list7[0:5:2]) #스텝을 2로 설정
print(list7[::-1]) #스텝을 -1로 설정: 리버스와 같은 기능
#step이 -1일 경우 start의 기본겂은 -1 stop의 기본값은 0으로 설정
print(list7[1::-1])
print(list7[:0:-1])
```

```python
list7=[100,5,70,40,50]
#sort(): 리스트의 요소를 오름차순으로 정렬
list7.sort()
print(list7)

#sort(reverse=True): 리스트의 요소를 내림차순으로 정렬
list7.sort(reverse=True)
print(list7)

#인플레이스 메소드... 값이 바뀐 것을 그대로 저장해 버린다
```

```python
list8=['apple', 'Apple.', 'banana', 'orange', 'melon']
print(list8)
list8.sort()
print(list8) #아스키코드 값으로 정렬
```

```python
list9=['김사과','오렌지','반하나','이메론']
list9.sort()
print(list9)
```

```python
list9 = [10,20,30,40,50,60,20,30,20]
#count(): 리스트에서 특정 요소의 갯수 반환
#숫자뿐만 아니라 문자도 가능
print(list9.count(20))
print(list9.count(100))
```
