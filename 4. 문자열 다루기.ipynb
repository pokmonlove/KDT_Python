#**1. 문자열이란**

파이썬에서 문자열은 텍스트 데이터를 다루는 데 사용되는 중요한 데이터 타입입니다. 문자열은 작은 따옴표(')나 큰 따옴표("), 세 개의 따옴표(''', """)로 둘러싸인 텍스트로 표현됩니다.



```python
str1 = '오늘도 신나는 파이썬 수업'

print(str1)

str2="오늘도 즐거운 파이썬 수업"
print(str2)

str3='''오늘도
재미있는
파이썬
수업
'''
print(str3) #문자의 서식이 그대로 출력된다
```

```python
str1 = 'hello'
str2="hello"

#is: 파이썬에서 두 객체가 같은 객체인지 여부 비교, 동일한 메모리 주소를 가지는지 확인
print(str1 is str2)
```

```python
#id(): 객체의 고유한 메모리 주소를 반환
text1='Hello'
print(id(text1)) #물리적 주소는 아님
text2='Hello'
print(id(text2))
text3='hello'
print(id(text3))
```

```python
a = 1000
b = 1000
print(a == b)
##결과: True
print(a is b)
#결과: False
print(id(a), id(b))
#결과: 134850552326736 134850552326096
```

#**3. 문자열 인덱싱과 슬라이싱**

문자열 내부의 문자에 인덱스를 사용하여 접근하거나 슬라이스할 수 있습니다.


###3-1. 인덱싱
인덱스를 이용해 문자를 추출한다 (0부터)

```python
text='문자열 내부의 문자에 인덱스를 사용하여 접근하거나 슬라이스할 수 있습니다.'
    # 0 1 2 3 4...
```

```python
print(text)
```

```python
print(text[0])
print(text[1])
print(text[12])
print(text[-1]) #마지막 문자
print(text[-2]) #마지막에서 두번째 문자
```

###3-2. 슬라이싱
문자열의 일부(구간)를 추출하는 방법

```python
print(text[0:6]) #0부터 6직전까지 (6포함 x) >> 0부터 5까지
```

```python
print(text[8:16]) #8부터 15까지
```

```python
#문자열 내부의 문자에 인덱스를 사용하여 접근하거나 슬라이스 가능

print(text[0:8])
print(text[8:16])
print(text[0:16])
print(text[8:])
print(text[:])
```

#**4. 문자열 메서드**

문자열 메서드는 문자열 객체에 적용되는 함수로, 문자열을 조작하거나 변환하는 다양한 작업을 수행할 수 있게 해줍니다.

```python
#len(문자열) 문자열.함수()
#upper() lower() 문자열을 대문자, 소문자로 변환
text="Hello Python"
print(text.upper())
print(text.lower())
```

```python
#count() 문자열의 특정 문자열의 등장 횟수를 반환

print(text.count('o'))
print(text.count('N'))
print(text.count('ll'))
print(text.count('l'))
```

```python
#find() 특정 문자열이 첫 번째로 등장하는 위치 (index) 반환
print(text.find('ll'))
print(text.find('LL')) #없으면 -1 반환
print(text.find('o'))
print(text.rfind('o'))
print(text.find("o",6)) #인덱스 6이후부터 찾는다
```

```python
#replace(): 문자열에서 특정 무자열을 찾아 다른 문자열로 변환
text = "hello python"
print(text.replace("python","world"))
```

```python
#strip(): 문자열의 양쪽 공백을 제거
text='                  hello python'
print(text)
print(text.strip())
```

```python
print('apple' == 'apple ')
print('apple' == 'apple'.strip())
```

```python
#split(): 특정 문자열을 기준으로 문자열 분리
text ='김사과 반하나 오렌지 이메론'
names=text.split() #공백을 기준으로 분리, 문자열을 함수에 전달하지 않으면 ' '기본값
print(names)
```

```python
text='원숭이1🍌원숭이2🍌원숭이3🍌원숭이4'
names=text.split('🍌')
print(names)
```

```python
text = '김사과, 반하나, 오렌지, 이메론'
names=text.split(',') #공백까지 같이 나뉘어져 버린다
names2=text.split(', ')
print(names)
print(names2)
```

```python
#join(): 문자열을 구성하는 각 요소 사이에 특정한 구분자를 삽입
text='hello'
print('🧏‍♀️'.join(text))
```

```python
alpha='abcde'
alpha='.'.join(alpha)
print(alpha)
alpha=alpha.split('.')
print(alpha) #문자를 따로따로 저장할 수 있는 방법
```

```python
#startswith(), endswith 문자열에서 특정 문자욜로 시작, 끝나는지 확인

text='Hello python'
print(text.startswith('Hello'))
print(text.startswith('hello'))
print(text.endswith('python'))
print(text.endswith('Python'))
```
