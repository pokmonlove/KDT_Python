###문제
score 리스트에 저장된 점수가 60점 이상인 갯수가 몇 개인지 확인 하는 프로그램

* score= [90, 40, 50, 60, 60, 70, 100, 40, 20]

```python
score= [90, 40, 50, 60, 60, 70, 100, 40, 20]
num=0
for i in range (len(score)):
  if score[i]>=60:
    num+=1
print(num)
```

```python
score= [90, 40, 50, 60, 60, 70, 100, 40, 20]
num=0
for i in score:
  if i>=60:
    num+=1
print(f"60명 이상인 학생의 수는 {num}명입니다.")
```

#딕셔너리와 for문

```python
dict1={'no':1,'userid':'apple', 'name':'김사과', 'hp':'010-1234-5678'}
print(dict1)

for i in dict1:
  print(i, end=' ')  #키만을 가져와 출력한다
```

```python
#dict1.keys()
for i in dict1.keys():
  print(i, end=' ') #위와 동일한 코드
```

```python
#value도 출력 가능
dict1={'no':1,'userid':'apple', 'name':'김사과', 'hp':'010-1234-5678'}
for i in dict1:
  print(dict1[i],end=' ')
```

```python
for i in dict1:
  print(dict1.get(i),end=' ') #get 메소드는 값이 없으면 에러가 나지 않는다
```

```python
for i in dict1.items():
  print(i) #튜플로 출력
```

```python
for k, v in dict1.items():
  print(k, v) #각각의 변수에 담아 출력
```

# **3. 컴프리헨션 (Comprehension)**

컴프리헨션은 파이썬에서 리스트, 세트, 딕셔너리 등의 컬렉션을 간단하게 생성하거나 변형하는 방법 중 하나입니다. 컴프리헨션은 반복문과 조건문을 사용하여 간결하게 컬렉션을 생성하는 기법으로, 코드를 더 간단하고 가독성 좋게 작성할 수 있도록 도와줍니다.



###3-1. 리스트 컴프리헨션

```python
num=10
result=[0 for i in range(num)] #10바퀴를 돌며 0을 반환해 준다
print(result) #0으로 10번 채워진 list 생성
```

```python
num=10
result=[i for i in range(num)]
print(result) #
```

```python
list1 =[10,20,30,40,60,70,80,90,100]
result=[num*2 for num in list1] #모든 값에 2를 곱하여 생성
print(result) #리스트 copy 메소드도 있기는 하지만 이런 방식도 있다
```

```python
list2=['apple','banana','orange','melon']
result=[len(str1) for str1 in list2]

print(result)
```

```python
result=[num for num in range(10) if num%2==0] #우측에다 조건 추가 #짝수 #반복에다 조건을 주게 된다
print(result)
```

```python
list3=[-1, 0, -5, 25, 5, -10, 2, 20]
#양수만 리스트에 그대로 저장하고 음수는 0으로 변환하여 저장 #조건을 주어 변환한다

result = [num if num>0 else 0 for num in list3]
print(result)
```

```python
###이 코드와 똑같다
list3 = [-1, 0, -5, 25, 5, -10, 2, 20]
result = []

for num in list3:
  if num > 0:
    result.append(num)
  else:
    result.append(0)

print(result)
```

```python
list4=[]
for i in range(1,4):
  for j in range(1,3):
    list4.append(i*j)

print(list4)
```

```python
list4=[i*j for i in range(1,4) for j in range(1,3)]

print(list4) #위 코드와 결과는 동일하다
```

###3-2. 리스트 컴프리헨션

```python
num = [1,2,3,4,4,5,5,5,6,6]
unique_num=set(num)
print(unique_num)#set을 활용해 중복 제거
```

```python
num = [1,2,3,4,4,5,5,5,6,6]
unique_num={i for i in num}
print(unique_num)
```
