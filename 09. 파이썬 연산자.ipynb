#**1. 산술 연산자**
산술 연산자는 주로 수치 데이터 유형에 대한 기본 산술 연산을 수행하기 위해 사용됩니다.


```
+ : 더하기
- : 빼기
* : 곱하기
/ : 나누기
//: 몫 (정수 나누기)
% : 나머지
**: 지수 (거듭제곱)
```



````python
num1=10
num2=3

print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)  #파이썬에서 정수 / 정수는 실수
print(num1//num2) #몫
print(num1%num2)  #나머지
print(num1**num2)
````

#**2. 비교 연산자**
비교 연산자는 주로 두 값을 비교하는 데 사용되며, 그 결과는 항상 불리언(True 또는 False) 값입니다.
```
==: 동등
!=: 동등하지 않음
< : 작다
<=: 작거나 같다
> : 크다
>=: 크거나 같다
```

````python
num1 = 30
num2 = 3

print(num1 > num2)
print(num1 >= num2)
print(num1 < num2)
print(num1 <= num2)
print(num1 == num2)
print(num1 != num2)
````

#**3. 할당연산자**

변수에 값을 할당하는데 사용됩니다. 파이썬에서는 기본 할당 연산자 외에도 복합 할당 연산자를 제공하여 코드를 간결하게 작성할 수 있게 도와줍니다.
```
= : 할당
+= : 더하기 후 할당
-= : 빼기 후 할당
*= : 곱하기 후 할당
/= : 나누기 후 할당
//=: 몫 후 할당
%= : 나머지 후 할당
**=: 지수 후 할당
```



````python
num=10
num +=1
print(num)

num-=1
print(num)

num*=10
print(num)

num/=10
print(num)

num %=2
print(num)
````

#**4. 비트 연산자**
비트 연산자는 정수를 이진 비트로 표현했을 때 사용하는 연산자들입니다. 각 연산자는 정수의 비트 단위로 동작합니다.

```
& : 비트 단위 AND
| : 비트 단위 OR
^ : 비트 단위 XOR
~ : 비트 단위 NOT
<< : 왼쪽 시프트
>> : 오른쪽 시프트

```

🥞 데이터의 크기를 나타내는 단위
1. 비트(bit) 데이터의 가장 작은 단위 (0또는 1의 값을 가질 수 있는 단위)
2. 바이트(byte) 8비트로 구성, 문자 하나(영문 기준)을 저장할 수 있는 크기
3. 킬로바이트(KB): 1024 byte
4. 메가바이트(MB): 1024 KB
5. 기가바이트(GB): 1024 MB
6. 테라바이트(TB): 1024 GB
7. 페타바이트(PB): 1024 TB
8. 엑사바이트, 제타 바이트, 요타바이트

비트: 컴퓨터의 용량 최소 단위

````python
num1 = 60 #00111100 #맨앞에 있는 0: 양수 #맨앞에 있는 1: 음수
num2 = 13 #00001101

print(num1&num2) #and 연산
print(num1|num2) #or 연산
print(num1^num2) #xor 연산
print(~num1) #not 연산 #부호를 반대로 바꿔준다 #11000011
#2의 보수로 음수를 구함
#비트를 모두 뒤집은 후, 1을 더한 값이 음수를 표현하는 방법
#제일 쉽게 생각하는 건 그냥 +1 한 후 부호를 바꾸면 된다
print(num1 <<2) #111100 + 00 = 11110000 --- 10진수로는 240
print(num1 >>2) #111100 - 00 = 1111 --- 10진수로는 15
````

#**5. 논리 연산자**

파이썬에서의 논리 연산자는 주로 불린(Boolean) 값을 조작하기 위해 사용됩니다. 그러나 파이썬의 논리 연산자는 다른 언어와는 다르게, 연산의 결과로 항상 True 또는 False만을 반환하는 것은 아닙니다. 대신 마지막으로 평가된 피연산자의 값을 반환합니다.



```
and: 논리 AND
or : 논리 OR
not: 논리 NOT
```

파이썬에서 불린으로 평가되는 값
1. True로 평가되는 값
```
Boolean: True
0이 아닌 숫자: -1, 1, 3.5 등
문자열과 바이트 문자열: 'Hello', b'hello'
모든 리스트, 튜플, 세트, 딕셔너리 (단, 빈 컨테이너가 아닌 형태)
```

2. False로 평가되는 값
```
Boolean: False
문자열: 빈 문자열'' 및 빈 바이트 문자열 b''
리스트, 튜플, 세트, 딕셔너리의 빈 값
None
```

````python
#값이 있는지 없는지로도 사용이 가능하다
````

````python
#and: 둘 다 true여야만 결과가 true
#or: 둘 중 하나여도 true면 결과가 true
#not: 반대

````

````python
num1 =10
num2 = 3

result1=num1> num2 #true
result2=num1==num2 #false

print(result1,', ', result2)
#and: 앞의 식의 결과가 true고 뒤의 식의 결과가 true일 때 결과가 True인 연산
print(result1 and result2)
#or: 앞의 식의 결과가 true이거나 뒤의 식의 결과가 true일 때 결과가 True인 연산
print(result1 or result2)
#not: 식의 값을 반대 부호로 변환
print(not result1)
````

````python
print(True and False)
print(True or False)
print(True and True)
print(not False)
````
