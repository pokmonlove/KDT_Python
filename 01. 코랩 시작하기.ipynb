#**4. 셀의 종류**

* 텍스트 셀: 마크다운 문법을 사용하여 텍스트를 작성할 수 있습니다.
* 코드 셀: Python 코드를 작성하고 실행할 수 있습니다.



## 4-1. 셀의 삽입 방법
* 상단의 +코드 또는 +텍스트 버튼을 누르면 삽입 가능
* 코드 셀을 위에 삽입: 컨트롤(커멘드)+M+A
* 코드 셀을 아래 삽입: 컨트롤(커멘드)+M+B


##4-2. 셀을 변경하는 방법
* 코드 셀을 텍스트 셀로 변경: ctrl+M+M
* 텍스트 셀을 코드 셀로 변경: ctrl+M+Y

##4-3. 셀을 삭제하는 방법
* 셀 우측의 휴지통 아이콘 클릭
* ctrl + M + D


#**5. 셀의 실행**
* ctrl + enter: 해당 셀을 실행하고 해당 셀에 커서 남음
* shift + enter: 해당 셀을 실행하고 다음 셀로 커서가 넘어감
* alt + enter: 해당 셀을 실행하고 다음 셀을 삽입(새로 생성)후 넘어감
