# *1. 튜플*
여러 가지 데이터 타입의 값을 저장할 수 있는, 변경 불가능(immutable)한 순서가 있는 컬렉션입니다

```python
'''
num=1
num2=2
...

...
num100=100
이렇게 많은 데이터를 한꺼번에 관리가 어렵기 때문에

'''
```

```python
tu1=(1) #튜플로 안 느끼고 계산을 위한 소괄호로 생각할 수도 있다
print(tu1)
print(type(tu1)) #그래서 int라고 나오는 것
```

```python
tu1=(1,) #튜플의 데이터가 하나일 때 꼭 콤마를 쓴다
print(tu1)
print(type(tu1))
```

```python
tu2=(1,3,5,7)
print(tu2)
print(type(tu2))
```

튜플과 리스트 자유자재로 변환이 가능해야 한다

```python
tu3=tuple([1,3,5,7])
print(tu3)
print(type(tu3))
```

```python
list3=list((1,3,5,7))
print(list3)
print(type(list3))
```

```python
tu4 = 1, 3, 5, 7
print(tu4)
print(type(tu4))
```

```python
num1 = 10
num2 = 20
num3 = 30
tu4= num1, num2, num3
print(tu4)
print(type(tu4))

result1 , result2, result3 = tu4 #다시 풀어줄 수 있다
print(result1)
print(result2)
print(result3)
```

```python
tu5 = ('apple', 'banana', ('🍎','🥭'))
print(tu5)
print(type(tu5))
print(tu5[0])
print(type(tu5[0]))
print(tu5[2][1])
print(type(tu5[2][1])) #이 또한 유니코드이므로 문자열이다
```

```python
print(2 in [2,4,6,8,10]) #컬렉션에 포함되어 있는 요소인지 확인해 준다
print(2 in (2,4,6,8,10)) #튜플에서도 확인이 가능하다
print(5 in (2,4,6,8,10))
```

```python
tuple9= (10, 30, 100, 90, 50)
#tuple9.sort() #AttributeError: 'tuple' object has no attribute 'sort'
#튜플은 값을 바꾸지 못하기 때문에 sort 메소드가 존재하지 않음
#sorted를 써야 하는데, 바뀐 값을 반환하기 때문에 반환값을 받을 변수를 만들어야 한다
result=sorted(tuple9)
result2=sorted(tuple9, reverse=True)
print(result)
print(type(result)) #결과가 리스트로 나온다 -- sorted 함수가 반환형을 리스트로 하기 때문에

result = tuple(result)
print(result)
print(result2)
```

```python
####문제 국어 점수 영어 점수 수학점수를 입력받아 총점, 평균을 구하는 프로그램 작성
#단, 입력 방식은 국/영/수 형태로 입력 #예시: 100, 75, 85
```

```python
kor, eng, math =input("국영수 점수를 / 단위로 입력하세요: ").split("/")
kor=int(kor)
eng=int(eng)
math=int(math)
total=kor+eng+math
avg=total/3
print(f"총점:{total}, 평균: {avg:.2f}")

#print(f"총점: {int(kor)+int(eng)+int(math)}")
#print(f"평균: {(int(kor)+int(eng)+int(math))/3}")
```
