#**1. 파일 입출력**

파이썬에서 파일 입출력은 다양한 용도로 사용됩니다. 텍스트 파일, 바이너리 파일, CSV 파일, JSON 파일 등을 읽고 쓸 수 있습니다.

### 1-1. 파일 열기

```
f = open("파일명", "모드")
파일명: 열고자 하는 파일의 이름이나 경로
모드: 파일을 어떻게 열 것인지를 지정

r: 읽기 모드 (기본값)
w: 쓰기 모드 (파일이 있으면 덮어쓰기)
a: 추가 모드 (파일의 끝에 내용을 추가)
b: 바이너리 모드 (텍스트가 아닌 바이너리 데이터를 읽고/쓸 때 사용)
t: 텍스트 모드
+: 읽기와 쓰기 모드
```

1-2. 파일 쓰기

````python
f=open("example.txt", 'wt') #쓰기 모드에 텍스트 모드 #t는 생략이 가능해서 'w'라고 써도 된다
````

````python
f.write("hello 파이썬\n")
f.close() #파일을 닫아야 저장이 완료됨
````

````python
f=open("example2.txt", 'w')
f.writelines(['라인1\n','라인2\n','라인3\n']) #리스트 형식으로 여러 줄을 쓸 수 있다
f.close()
````

````python
f = open("example3.txt",'w')
for i in range(10):
  f.write("파일 쓰기 테스트"+str(i)+'\n')
f.close()
print("파일 쓰기 완료")
````

# **2. with문 사용하기**

* 파일을 보다 안전하고 깔끔하게 다루는 방법 중 하나
* 컨텍스트 매니저를 사용하여 파일을 열고 파일 작업이 끝나면 자동으로 파일을 닫아주는 역할
* 파일을 열고 나서 오류가 발생하면 오류가 발생하더라도 파일을 자동으로 닫아주는 역할
* close() 호출하지 않아도 되므로 코드가 간단해지고 깔끔해짐

````python
with open("word.txt",'w') as f:
  while True:
    data= input("단어를 입력하세요")
    if data.lower()=='quit':
      break
    f.write(data+"\n")
````

````python
with open("example3.txt",'rt') as f:
  while True:
    data = f.read(5)
    if not data:#""
      break
    print(data, end="")
````

````python
with open('word.txt','r') as f:
  lines=[]
  while True:
    line = f.readline() #한 줄만 읽어 온다
    if not line:
      break
    print(line, end='')
    lines.append(line.strip())
````

````python
print(lines)
````

````python
with open("word.txt",'r') as f:
  lines=f.readlines()
  print(lines)
````

````python
li2=[]
for i in lines:
  li2.append(i.replace('\n',''))
print(li2)
````
