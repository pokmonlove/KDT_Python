#**3. else 문**
else는 파이썬의 조건문 구조에서 if와 elif 조건들이 모두 거짓으로 평가될 경우 실행되는 부분을 정의하는 데 사용됩니다. else는 선택적으로 사용될 수 있으며, 그 자체로는 어떤 조건을 갖지 않습니다. 다시 말해, else 블록은 위의 if와 elif 조건들이 모두 만족되지 않는 경우에만 실행됩니다.

```
if 조건1:
    코드 블록1
elif 조건2:
    코드 블록2
else:
    코드 블록3
 ```

````python
age=int(input("나이를 입력하세요: "))
if age > 19:
  print("성인입니다")
  print("넷플릭스 성인 영화 감상 가능")
else:
  print("미성년자입니다")
  print("디즈니 플러스 보세용")
print("입력한 나이는 %d살입니다" %age)
````

#**4. elif 문**
elif는 if-else 구조에서 추가적인 조건을 검사하기 위해 사용되는 키워드입니다. elif는 "else if"의 줄임말이며, 주어진 여러 조건 중에서 하나만 참일 때 실행되는 코드 블록을 정의합니다.

```
if 조건1:       #True: 조건 1 실행, 밑에는 쳐다도 안 보고 나감. False: 조건 2 실행
    코드 블록1
elif 조건2:  
    코드 블록2
elif 조건3:
    코드 블록3
else:
    코드 블록4
 ```

True에 해당하는 문장을 실행하고, 아래 코드는 보지도 않은 채 바로 밖으로 나간다

````python
age=int(input("나이를 입력하세요: "))
if age > 19:
  print("성인입니다")
elif age>14:  #15살 이상, 19살 이하
  print("청소년입니다")
elif age>6:  #7살 이상, 14살 이하
  print("어린이입니다")
elif age>0:  #1살 이상, 6살 이하
  print("유아입니다")
else:
  print("다시 입력해 주세요")
print("입력한 나이는 %d살입니다" %age)
````

* elif는 if 없이 단독으로 사용될 수 없습니다.
* elif는 필요에 따라 여러 번 사용될 수 있습니다.
* else는 선택적입니다. 사용될 경우, 항상 if나 elif 뒤에 나타나야 합니다.


elif를 사용함으로써 여러 조건을 순차적으로 검사하고, 해당 조건에 따라 다른 행동을 취할 수 있습니다. 이는 코드의 가독성을 높이고, 조건의 우선순위를 명확하게 만들어 줍니다.

#**5. 조건부 표현식(Conditional Expression)**

조건부 표현식은 파이썬에서 삼항연산자라고도 불립니다. 간단한 if-else 구문을 더 간결하게 표현할 수 있는 방법을 제공합니다.

```
조건이 참인 경우 값 if condition else 조건이 거짓인 경우 값
```

참일 경우 실행 **if** 조건식 **else** 거짓일 경우 실행

````python
num = int(input('숫자를 입력하세요: '))

if num % 2 == 0:
    print('짝수')
else:
    print('홀수')
````

###문제 1

1부터 100까지 짝수의 합을 출력하는 프로그램을 작성해 보자

````python
i=1
sum=0
while i<=100:
  if i%2==0:
    sum+=i
  i+=1
print(sum)
````

###문제 2.

원하는 단을 입력받아 해당 단을 구구단으로 출력해 보자

* 원하는 단 입력하세요: 입력받고
```
3단
3*1=3
3*2=6...
3*9=27
```

````python
num=int(input("원하는 단 입력하세요: "))
print(f"**{num}단**")
i=1
while i<10:
  print(f"{num}*{i}={num*i}")
  i=i+1
````

````python
#짝수인 경우 종료를 시키고, 홀수인 경우 계속 숫자를 입력받는 프로그램
#break: 반복문을 즉시 종료할 때 사용되는 키워드
#while 또는 for문에서 모두 사용
while True:
  num=int(input("짝수 입력"))
  if num%2==0:
    print("짝수입니다")
    print("프로그램을 종료합니다.")
    break #이 반복문을 나가라는 뜻입니다. 프로그램 종료가 아님
  else:
    print("입력한 숫자가 홀수입니다. 다시 입력해 주세요")

print("이 메시지가 보인다면 성공!")
````
