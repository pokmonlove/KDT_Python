#**1. 스코프(scope)**
파이썬에서 변수의 범위(scope)는 해당 변수가 프로그램 내에서 참조되고 변경될 수 있는 영역을 의미합니다.

즉, 변수를 쓸 수 있는 범위

### 1-1. 지역 변수(Local)

Local scope는 변수가 함수 내에서 정의될 때 해당 함수 내부를 의미합니다.

````python
def local_func():
  local_var = '로컬 변수'
  print(local_var)

#로컬 변수의 생명은 이 함수에서만
````

````python
local_func()
````

````python
#print(local_var) #NameError: name 'local_var' is not defined
````

###1-2. Enclosing(둘러싼 범위):
내부 함수에서 외부 함수의 변수를 참조할 때 해당 외부 함수 범위를 의미합니다.

````python
def outer_func():
  enclosing_var="둘러싼 범위 변수"
  #함수 안에 함수를 만들 수도 있다
  def inner_func():
    print(enclosing_var)
  inner_func()
````

````python
#inner_func() #NameError: name 'inner_func' is not defined
#안의 함수를 부를 수는 없다
````

````python
outer_func() #안쪽의 함수를 가져올 수 있다
````


###1-3. Global(전역 범위)

변수가 스크립트의 최상위 수준에서 정의될 때 전체 스크립트 파일 내에서 해당 변수를 의미합니다.

````python
num = 2 #전체 스크립트 파일 내에서 해당 변수를 의미한다
````

````python
global_var="글로벌 변수"
def test_func():
  print(global_var)
````

````python
test_func() #아예 바깥에 있는 변수 사용
````

````python
global_var = 10 #전역 변수
def test_func():
  global_var=20 #지역 변수
  print(global_var)

test_func()
print(global_var)
````

````python
#전역변수의 값을 수정하고 싶어! > global 키워드 사용
````

````python
global_var = 10 #전역 변수
def test_func():
  global global_var
  global_var=20 #전역 변수의 값을 수정
  print(global_var)

test_func()
print(global_var)
````


###1-4. Built-in(내장 범위)

파이썬의 내장 함수와 모듈들이 속하는 범위입니다. 예를 들어, print(), len() 같은 함수들이 여기에 해당됩니다.

#**2. 변수의 범위 탐색 순서**

```
1. Local scope
2. Enclosing scope
3. Global scope
4. Built-in scope
```
 지역 범위에 동일한 이름의 변수가 없으면 파이썬은 둘러싼 범위를 확인하고, 그 다음으로 전역 범위, 마지막으로 내장 범위를 확인합니다.

이렇게 변수의 범위를 이해하고 관리하는 것은 코드의 가독성과 유지 보수성을 높이고, 예기치 않은 오류를 방지하는 데 중요합니다.

함수의 이름이 겹치게 되면 저 순서로 작동한다.
