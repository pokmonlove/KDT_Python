#**1. 사용자 정의 함수**
사용자 정의 함수란 사용자가 특정 작업을 수행하기 위해 직접 작성한 함수를 의미합니다. 파이썬에는 많은 내장 함수들이 있지만, 때로는 우리의 요구사항에 맞게 동작하는 함수를 직접 만들어야 할 때가 있습니다. 이때 사용자 정의 함수를 작성하게 됩니다. 기본적인 구조는 아래와 같습니다.

```
def 함수명(매개변수1, 매개변수2, ...):
    # 함수 내용
    return 결과값
```

##1-1. 매개변수와 반환값이 없는 함수

````python
def func1():
  print("처음으로 만드는 함수")
````

````python
func1()
func1()
````

````python
for i in range(5):
  func1()
````

##1-2. 매개변수가 있고 반환값이 없는 함수

````python
def func2(num):
  print(f"입력받은 숫자:{num:}")
````

````python
func2(100)
func2(5)
````

````python
def func3(start, end):
  sum=0
  for i in range(start, end+1):
    sum+=i
  print(f"{start}부터 {end}까지의 합: {sum}")
````

````python
func3(1,10)
func3(1,100)
````

##1-3.반환값이 있는 함수

return 키워드 등장

````python
def func1():
  print("처음으로 만드는 함수")
````

````python
func1()
temp=func1() #반환값을 한번 찍어보자
print(temp) #return이 없어서 None
````

````python
temp=func1 #괄호 없이 사용
print(temp) #이 함수가 저장되어 있는 메모리 주소를 가리키게 되었다
#<function func1 at 0x7dbe10c52050>
temp() #func1을 호출한 것과 같다
````

````python
def func4():
  return '🎁'
````

````python
func4() #문자열이 return 되는 것을 볼 수 있다
````

````python
temp=func4
temp()

print(temp()) #temp가 func4함수의 주소를 가리키게 되어 결국 func4와 같은 동작을 한다
print(type(temp))
````

````python
temp=func4()
print(f"temp에 저장된 값: {temp}")
print(type(temp))
````

##1-5. 가변 매개변수
함수를 호출할 때 *를 사용하면 시퀀스(리스트, 튜플 등)의 요소를 개별적인 위치 인자로 풀어서 전달할 수 있습니다.


````python
def func7(*args):
   return args
````

````python
'''
def func7(*args):
   for i in args:
    print(i)

func7(1,2,3,4)
'''
````

````python
list_a=[1,2,3,4]
print(func7(list_a))
print(func7(*list_a))
````

````python
print(func7())
print(func7(1,2,3,4))
print(func7([1,2,3,4]))
print(func7({1,2,3,4}))
````

````python
def add(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
print(add(*numbers))  # 출력: 6
````

````python
def print_numbers(*args):
  for arg in args:
    print(arg, end=" ")
````

````python
print_numbers(10)
print_numbers(10, 20, 30, 40)
````

````python
list1=[10, 20, 30, 40]
print_numbers(list1)
print(" ")
print_numbers(*list1) #두 결과가 다르다
````

````python
list1=[10, 20, 30, 40]
list2=[30, 40, 50, 60]
print_numbers(*list1,*list2)
print(" ")
print_numbers(*list1,list2)
print(" ")
print_numbers(list1,list2)
````

````python
def sum_num(*args):
  sum=0
  for arg in args:
    sum+=arg
  return sum
````

````python
list1=[10, 20, 30, 40]
#sum_num(list1)  #TypeError:
sum_num(*list1)
````

##1-6. 키워드 매개변수

````python
def func8(id, name, age):
  print(f'아이디: {id}')
  print(f'이름: {name}')
  print(f'나이: {age}')
````

````python
func8("apple","이사과",20)
````

````python
dic1={'age':25, "id":"banana","name":"바나나"}
func8(**dic1) #키에 맞춰 매개변수를 찾아가 세팅이된다
````

````python
func8(*dic1) #*를 하나만 쓰면 키 값만 뽑아서 간다
````

##1-7. 여러 개의 반환값

````python
def func6(num1=0, num2=0):
  sum=num1+num2
  return sum
````

````python
result1=func6(10,3)
print(result1)
#덧셈뿐만 아니라 사칙연산을 전부 구하고 싶다면?
````

````python
def func9(num1=0, num2=0):
  sum=num1+num2
  return sum, num1-num2, num1*num2, num1/num2
````

````python
result1, result2, result3, result4 = func9(10,3)
print(f'덧셈: {result1}')
print(f'뺄셈: {result2}')
print(f'곱셈: {result3}')
print(f'나눗셈: {result4:.2f}')
````

````python
#곱셈만 쓰고 싶다면
_,_,result3,_=func9(10,3) #변수를 만들지 않아 메모리를 아낄 수 있다
print(f'곱셈: {result3}')
````

##**None의 특징**

None은 파이썬에서 특별한 값으로, 아무런 값이 없음을 표현하는 데 사용됩니다. 다른 언어에서의 null 또는 nil과 유사한 개념입니다.

None은 파이썬의 내장 상수이며, 그 자체로 데이터 타입이 NoneType입니다. 모든 None은 동일하므로, 두 개의 None 값을 비교할 때 항상 True를 반환합니다.

````python
num=None
print(num)
print(type(num))
````

````python
def func10(num):
  if num>=10:
    return num
````

````python
print(func10(10))
print(func10(5))
````

````python
def func11(num=None):
  if num is None:
    print("전달된 값이 없습니다")
  else:
    print(num)
````

````python
func11(10)
func11()
````

````python
num = None
#None을 비교할 경유 == 보다 is를 권장한다
if num is None:
  print("num의 값이 없습니다") #그러나 이것을 사용하는 걸 추천

if num == None:
  print("num의 값이 없습니다") #이것도 작동한다

  #is: 주소를 비교, ==: 값을 비교
````

````python
str1="Hello"
str2="Hello"

#객체 비교
print(str1 is str2)

li1 = [1,2,3,4,5]
li2 = [1,2,3,4,5]

#객체 비교
print(li1 is li2) #객체가 따로 만들어지기 때문에 False #메모리 주소가 다르다

#값 비교
print(li1 == li2)

````
