# **1. print 함수**

print() 함수는 파이썬의 기본 출력 함수로, 터미널이나 콘솔 창에 텍스트나 데이터를 출력하기 위해 사용됩니다.

```python
print("안녕하세요 파이썬") #결과가 잘 나오는지 보기 위함으로 주로 쓰입니다
```

```python
print('안녕하세요 파이썬')
```

```python
print("'안녕하세요'")
```

```python
print(type(10))
print(type('10'))
```

```python
print(10,20,30,40,50)
```

```python
print('hi','python','nice',sep=' ')
print('hi','python','nice',sep='😎')
print('hi','python','nice',sep='*')
```

#**2. 파이썬의 escape 문자**

파이썬에서 이스케이프 문자는 문자열 안에서 특별한 의미를 가지는 문자들을 나타내기 위해 사용됩니다. 이스케이프 문자는 백슬래시(\)로 시작하며, 문자`열 내에서 특정한 동작이나 특수 문자를 표현하기 위해 사용됩니다.

```python
print("안녕하세요\n 파이썬")
print("이름: \t 김사과")
print("\\재미이따\\")
print("이것은 오늘의 \"날씨\"가 맑음")
```

#**3. print 함수 내의 연산**
print 함수는 문자열뿐만 아니라 숫자의 연산결과도 출력 가능


```python
print(10)
print(10+5)
print('덧셈: ', 10+5)
print("사칙연산: ", (10+5)*2)
print("나눗셈: ", 10 / 2) #유일하게 파이썬 연산자에서 정수%정수는 실수
print("정수 나눗셈: ", 10//3) #몫만 나온다
print("나머지: ", 10%3) #나머지
print("거듭제곱: ", 10**3) #10의 세제곱
```

#**4. 출력 서식 지정하기**


파이썬의 print() 함수는 출력 서식을 지정하여 값을 형식화하고 출력하는 기능을 제공합니다. 서식 지정자와 함께 사용되는 % 연산자와 format() 메서드를 사용하여 원하는 형식으로 출력할 수 있습니다

```python
print("정수:",100)
print("정수:%d" % 100) #decimal 정수
print("정수:%d" %100.5)
print("실수: %f" %10.99)
print("실수:%.5f"%100)
print("문자열: %s" %'파이썬')
print("문자열: %s" %100)
```

```python
#fstring
print(f'정수: {100}')
```

```python
#중괄호 안의 변수를 계속 수정하며 출력하는 서식 방법
num=100
print(f'정수: {num}')
print("정수: %d " %num)
```

```python
print(f"정수 연산:{100+50}")
```

```python
print(f"정수 연산:{100+50}")
print('5+3=8')
print('%d+%d=%d' %(5,3,5+3)) #튜플 형태로 서식해야 함
print(f'{5}+{3}={5+3}')
```

```python
#format 메소드 #문자열에만 접근할 수 있는 메소드다
print('내친구들: {},{},{}'.format('바나나','오렌지','메론'))
```

```python
print('이름: %s, 나이: %d'%('김사과', 20))
```

```python
print('이름: %s, 나이: %d'%('김사과', 20))
print('몸무게: %f'%50.1234346)
print('몸무게" %.2f'%50.128124)
```

#**5.주석문**

파이썬 주석은 코드 내에 프로그래머가 작성한 설명이나 메모를 포함하는데 사용됩니다. 주석은 코드를 이해하고 다른 개발자와 협업하는 데 도움이 되며, 코드의 기능과 목적을 명확하게 설명할 수 있습니다. 주석은 프로그램 실행 중에 무시되므로 코드 실행에 영향을 미치지 않습니다.

한줄 주석

```python
print("파이썬 출력하기") #문자열 출력은 작은 따옴표 혹은 큰 따옴표를 사용한다 ##주석은 안 보이지롱
```

```python
'''
이곳은 여러 줄 주석
여러 줄 주석은 큰따옴표 세 개 혹은 작은 따옴표 세 개 사이에 작성합니다
여러 줄의 설명을 적을 수 있다
'''
print("파이썬 출력")
```
