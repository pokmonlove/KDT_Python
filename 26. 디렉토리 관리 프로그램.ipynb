# **1. 압축 파일 정리하기**

````python
import os
import glob
import zipfile
import shutil
import fnmatch
import pathlib
````

````python
#현재 경로 알아보기
os.getcwd()
````

````python
# 정리 대상 디렉토리 경로 설정
target_path='./정리'
````

````python
# 정리 디렉토리에서 압축 파일 확인하기
zipfile_path=[]
for filename in glob.glob(os.path.join(target_path, '**/*.zip'), recursive=True):
    zipfile_path.append(filename)
print(zipfile_path)
````

````python
# 압축 파일 해제하기
for filename in zipfile_path:
    with zipfile.ZipFile(filename) as myzip:
        myzip.extractall(target_path) #현재 자리
````

````python
# 파일 이름변경 후 압축파일 해제하기
for filename in zipfile_path:
    with zipfile.ZipFile(filename) as myzip:
        zipinfo=myzip.infolist()
        #print(zipinfo)

        for info in zipinfo:
            decode_name=info.filename.encode('cp437').decode('euc-kr')
            info.filename=os.path.join(target_path, decode_name)
            myzip.extract(info)
````

# **2. 파일명 정리하기**


````python
#!pip install openpyxl
import openpyxl as opx
````

````python
def getFileName(target_path):
    wb=opx.Workbook()
    ws=wb.active

    ws.cell(row=1, column=1).value="파일경로"
    ws.cell(row=1, column=2).value="파일명(변경 전)"
    ws.cell(row=1, column=3).value="파일명(변경 후)"

    i=2

    current_dir=target_path
    filelist=os.listdir(current_dir)

    for filename in filelist: # 파일 개수만큼 반복
        ws.cell(row=i, column = 1).value=current_dir+'/' # 파일 경로
        ws.cell(row=i, column = 2).value=filename        # 파일명
        i+=1

    wb.save(os.path.join(target_path, 'filelist.xlsx'))
````

````python
getFileName(target_path)
````

```

```
* 단 튜플로 디렉토리, 변경 전 파일명, 변경 후 파일 명)으로 나열되는 리스트를 반환
* 예) [('./정리', C_2022_09_10_데이터베이스_ERD_001 (1).xlsx, C_2022_09_10_데이터베이스_ERD_001.xlsx) ...]


````python
def exelRead(filepath):
    mylist=[]
    wb = opx.load_workbook(filepath)
    ws=wb.active
    num=len(ws['B'])
    for i in range(2, num+1):
        if (ws[f'B{i}'].value) != (ws[f'C{i}'].value):
             mylist.append((ws[f'A{i}'].value, ws[f'B{i}'].value, ws[f'C{i}'].value))
    wb.close() ## Workbook 종료
    print(mylist)
````

````python
def exelRead(filepath):
    mylist=[]
    wb = opx.load_workbook(filepath)
    ws=wb.active
    dirpath= [r[0].value for r in ws]
    file_before= [r[1].value for r in ws]
    file_after= [r[2].value for r in ws]

    datalist=[]
    #for t in zip(dirpath, file_before, file_after):
    #   datalist.append(t)
    len_num=len(dirpath)
    for i in range(1, len_num):
        temp_tuple=(dirpath[i], file_before[i], file_after[i])
        datalist.append(temp_tuple)

    return datalist
````

* 단, 카테고리가 없으면 빼고 생각
* 단, fnmatch.fnmatch() 메서드를 사용
* ['생산', '클래스설계', '물류', 'erd','인사']


````python
def categoryList(target_path):
    catelist=['생산', '클래스설계', '물류', 'erd','인사']
    for filename in os.listdir(target_path):
        for cate in catelist:
            if fnmatch.fnmatch(filename, f'*_{cate}_[0-9][0-9][0-9].*'):
                print(filename)
````

````python
def categoryList(target_path):
    file_list=[]
    for filename in os.listdir(target_path):
        if fnmatch.fnmatch(filename, '*_[0-9][0-9][0-9].*'):
            file_list.append(filename)
    category=[]
    for file in file_list:
        temp_list=file.split('_') #A 2022 06 30 생산로그 생산 002.pdf'
        category.append(temp_list[-2])
    category=set(category)
    return list(category)
````

### 문제
* 정리 디렉토리에 있는 모든 파일을 new_정리에 있는 디렉토리에 따라 파일을 분류해 저장
* 단 파일 이름에 따라 디렉토리에 맞춰 저장(카테고리가 맞지 않는 파일은 모두 '기타'에 저장)

```
def moveFile(new_path, target_path, catelist):
    pass
```

````python
target_path='./정리'
````

````python
def moveFile(new_path, target_path, catelist):
    dirlist=os.listdir(new_path)
    filelist=os.listdir(target_path)
    #print(dirlist, filelist)
    categorydic={}

    for file in filelist:
        try:
            temp_list=file.split('_')
            assert temp_list[-2] in catelist
            categorydic[file]=temp_list[-2]
        except:
            categorydic[file]='기타'
    #print(target_path)
    for key, value in categorydic.items():
        shutil.copy(target_path+ '/' + key, new_path +'/' +value)
````

````python
moveFile(new_path, target_path, catelist)
````

참고: assert 문
* 파이썬에서 디버깅 목적으로 사용하는 구문
* 조건이 참인지 확인하고, 조건이 거짓이라면 AssertionError 발생시킴
* 의도한 대로 프로그램이 작동하는지 확인하기 위한 용도
* 디버그 모드에만 동작하며, 배포시 동작이 되지 않는다

````python
def divide(a,b):
    assert b!=0, "B는 0이 될 수 없다"
    return a/b
````

````python
print(divide(10,2))
````

````python
print(divide(10,0)) #AssertionError: B는 0이 될 수 없다
````
