#**1. 세트**
세트(set)는 중복되지 않는 항목들의 순서가 없는 컬렉션입니다.

```python
s1 = {1, 3, 5, 7}
print(s1)
print(type(s1))
```

```python
s2={} #데이터를 안 넣었을 때 다른 컬렉션으로 인식해 버린다
#튜플은 () 상태여도 튜플이 만들어졌는데...
print(s2)
print(type(s2)) #무조건 데이터가 들어가야 한다
```

```python
s3 = set([1,4,5,7])
print(s3)
print(type(s3))
```

```python
s3 = set([1,7,1,5])
print(s3)
print(type(s3))
```

```python
list1 = [1, 2, 4, 5]
s4=set(list1)
print(s4)
print(type(s4))
```

```python
s5={1,4,5,6,5,6,5,6,10,4,5}
print(s5)
```

```python
list2=[1,3,4,5,6,9,4,5,6,8]
print(list2)
s6=set(list2)
print(s6) #set으로 바꾸었더니 중복된 데이터가 생략된다
```

#**2. 세트의 메서드**



```python
#add(): 세트의 요소를 추가
s1 = {1, 3, 5, 7, 9}
print(s1)

s1.add(2)
print(s1) #따로 크기에 따라 정렬되지 않는다.
#세트는 순서가 없는 컬렉션 타입이므로 아무데나 들어간 것

s1.add(5)
s1.add(10)
print(s1)

s1.add(7)
print(s1) #중복된 데이터는 들어가지 않는다
```

```python
s1 = {1, 3, 5, 7, 9}
print(s1)

#undate(): 세트에 여러 요소를 추가
s1.update([2,4,6,8])
print(s1)
```

```python
s1 = {1, 3, 5, 7, 9}
print(s1)

#remove(): 세트의 요소를 제거, 단 요소가 없다면 에러 발생
s1.remove(3)
print(s1)
#s1.remove(3) #KeyError : 3
#print(s1)
```

```python
#discard(): 세트의 요소를 제거, 단 요소가 없어도 에러가 발생하지 않는다
#지우든 말든 상관이 없을 때 사용한다

s1 = {1, 3, 5, 7, 9}
print(s1)
s1.discard(3)
print(s1)
s1.discard(3)
print(s1)
#remove()와 달리 에러가 발생하지 않는다
```

```python
s1 = {1, 3, 5, 7, 9}
print(s1)
#메모리 주소를 복사한다 - 다른 변수에서 수정을 하면 원본도 같이 변한다
s2=s1 #같은 곳을 가리킨다
print(id(s1))
print(id(s2))

#copy(): 세트를 복사한다. 변수가 다른 곳을 가리키게 된다
#값을 복사한다 - 메모리 주소는 다르다 - 다른 변수에서 수정해도 원본은 변하지 않는다
s2=s1.copy()
print(s1)
print(s2)
print(id(s1))
print(id(s2)) #서로 메모리 주소가 다르다
#
```

```python
#합집합
s3={10,20,30,40,50,60,70}
s4={30,40,50,60,80}

#union(): 합집합을 계산하여 반환
result1=s3.union(s4)
print(result1)
```

```python
result2=s3|s4  #| 기호를 사용하면 위와 똑같은 결과
print(result2)
```

```python
#교집합
s3={10,20,30,40,50,60,70}
s4={30,40,50,60,80}

#intersection(): 교집합을 계산하여 반환
result3=s3.intersection(s4)
print(result3)
```

```python
result4=s3&s4 ##& 기호를 사용하면 위와 똑같은 결과
print(result4)
```

```python
#차집합 구하기
s3={10,20,30,40,50,60,70}
s4={30,40,50,60,80}

#difference(): 차집합을 계산하여 반환
result1 = s3.difference(s4)
print(result1)
```

```python
result2=s3-s4 ##- 기호를 사용하면 위와 똑같은 결과
print(result2)
```

```python
#symetric_difference(): 대칭 차집합을 계산하여 반환 (양쪽의 차집합/전체-교집합)

s3={10,20,30,40,50,60,70}
s4={30,40,50,60,80}

result1 = s3.symmetric_difference(s4)
print(result1)

result2=s3^s4 ##^ 기호를 사용하면 위와 똑같은 결과
print(result2)
```
