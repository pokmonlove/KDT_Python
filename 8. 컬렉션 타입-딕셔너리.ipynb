#**1. 딕셔너리**
파이썬의 딕셔너리는 키-값 쌍을 저장하는 변경 가능한(mutable) 컬렉션입니다.

* 키는 중복되면 안 된다
* 값은 중복돼도 괜찮다

```python
dict1={}
print(dict1)
print(type(dict1))
```

```python
dict2={1:'김사과'}
#      키   값

dict2={1:'김사과',2:'반하나',3:'오렌지',4:'이메론'}
print(dict2)
print(type(dict))
```

```python
dict2={1:'김사과',2:'반하나',2:'오렌지',4:'이메론'}
#키가 중복되면 값을 재저장 한다 (x에러)
print(dict2)
print(type(dict))
```

```python
dict2={1:'김사과',2:'반하나',3:'오렌지',4:'이메론'}
print(dict2[1])
print(dict2[2])
print(dict2[3])
print(dict2[4])
```

```python
#키는 문자여도 괜찮다
dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)
print(dict3['no'])
print(dict3['hp'])
```

```python
dict4={1:'apple'}
print(dict4)
```

```python
dict4[100] = 'banana'
print(dict4) #없던 키를 넣으면 데이터가 추가된다

dict4[50] = 'orange'
print(dict4)
```

```python
#set처럼 순서가 없는 컬렉션 타입이다.
```

```python
dict4[100]='melon'
print(dict4) #키가 있으면 값을 변경한다
```

```python
del dict4[100]
print(dict4)
```

```python
dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)
#요소 추가
dict3['gender']='female'
print(dict3)
#요소의 값 변경
dict3['no']=10
print(dict3)
#딕셔너리의 값은 컬렉션 타입도 들어갈 수 있다
dict3['score']=[100,300,400]
print(dict3)

#키 값에 리스트가 들어갈 수 없다
#dict3[[1,2,3]]=[10,20,30] #TypeError: unhashable type: 'list'
#print(dict3)
#키 값에 컬렉션 타입이 들어갈 수 있긴 하다. 단 변경 불가능한 데이터(튜플 같은)가 들어가야 한다
dict3[(1,2,3)]=[10,20,30]
print(dict3)

#값에는 딕셔너리가 들어갈 수 있다
dict3['과일']={'사과':'🍎','오렌지':'🍊','바나나':'🍌'}
print(dict3)
```

#**2. 딕셔너리의 메서드**



```python
dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)
#keys(): 딕셔너리의 모든 키를 반환
print(dict3.keys())
#values(): 딕셔너리의 모든 값을 반환
print(dict3.values())
#items(): 딕셔너리의 모든 키-값 쌍을 튜플 형태로 반환
print(dict3.items())
```

```python
dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)
```

```python
#get(): 특정 키에 대한 값을 반환 만약 키가 딕셔너리에 없으면 None을 반환
#None을 치환할 수 있는 문자열을 설정할 수 있음
print(dict3.get('userid'))

#print(dict3['address']) #KeyError: 'address'
print(dict3.get('address')) #없는 키를 불렀는데도 에러 x
print(dict3.get('address','주소 알 수 없음')) #키가 없다면 대체하는 문자열이 나온다
print(dict3.get('name','이름 알 수 없음')) #키가 있으면 잘 나온다
print(dict3)
```

```python
num = ' '
num=None #가리키는 변수가 없다
```

```python
#pop():특정 키에 대한 값을 제거하고, 제거된 값을 반환
#키가 없다면 에러가 난다

dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)

temp=dict3.pop('hp')
print(temp) #키 값인 no가 아닌 값만 저장된다
```

```python
#in: 딕셔너리의 특정 키가 있는지 확인
dict3={'no':1, 'userid' : 'apple', 'name':'김사과','hp':'010-7777-7777'}
print(dict3)

print('hp'in dict3)
print('김사과'in dict3) #값이 있는지는 확인해 주지 않는다
```
