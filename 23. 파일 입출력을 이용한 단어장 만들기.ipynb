* 단 프로그램을 종료할 때까지 반복하여 메뉴를 입력받도록 하며 클래스를 이용하여 설계하도록 한다.
* 메뉴 입력시 1~5가 아닌 숫자가 입력되는 경우 "다시 입력하세요"라고 메시지 출력
(오류 처리로 한다)

`````python
#클래스 처리
class WordDict:
  def __init__(self):
    self.word=input("단어를 입력하세요")
    self.mean=input("뜻을 입력하세요")
    self.level=input("레벨을 입력하세요")
    print(">> 등록되었습니다")
`````

`````python
WordList=[]
MeanList=[]
levelList=[]

def func1():
  word1=WordDict()
  WordList.append(word1.word)
  MeanList.append(word1.mean)
  levelList.append(word1.level)
  #print(word1.word)

def func2():
  for i in range(len(WordList)):
    print(f"{WordList[i]}, {MeanList[i]}, (레벨 {levelList[i]})")

def func3():
  with open("word.txt",'w') as f:
    for i in range(len(WordList)):
      f.write(WordList[i]+MeanList[i]+levelList[i]+"\n")

def func4():
  with open("word.txt",'r') as f:
    li2=[]
    lines=f.readlines()
    for i in lines:
      li2.append(i.replace('\n',''))
    print(li2)
`````

`````python
def func():
  print("단어장 메뉴\
  1. 등록하기\
  2. 출력하기\
  3. 저장하기\
  4. 불러오기\
  5. 종료하기")
  select=input("번호를 입력해 주세요: ")
  if select=='1':
    print("1. 등록하기")
    func1()
    print("===============================")
    return True
  elif select=='2':
    print("2. 출력하기")
    func2()
    print("===============================")
    return True
  elif select=='3':
    print("3. 저장하기")
    func3()
    print(">> 저장되었습니다(문서생성)")
    print("===============================")
    return True
  elif select=='4':
    print("4.불러오기")
    func4()
    print("불러왔습니다")
    print("===============================")
    return True
  elif select=='5':
    print("5. 종료하기")
    print(">> 프로그램을 종료합니다.")
    return False
  else:
    raise Exception('다시 입력해 주세요')
`````

`````python
print("나만의 단어장 만들기")
while func()==True:
  try:
      func()
      if func()==False:
        break
  except Exception as e:
      print(e)
`````

`````python
class Word: #단어들을 변수에 담아 두면 따로 노니까 하나로 묶어 주고 싶어요 #ENG/Kor/Lev
  def __init__(self,eng,kor,lev=1):
    self.eng=eng
    self.kor=kor
    self.lev=lev

  def setEng(self, eng):
    self.eng=eng

  def getEng(self):
    return self.eng

  def setKor(self, kor):
    self.kor=kor

  def getKor(self):
    return self.kor

  def setLev(self, lev):
    self.lev=lev

  def getLev(self):
    return self.lev
`````

`````python
wordList=[] #전역 리스트
filename="word.txt"
`````

`````python
class WordService: #메소드를 만들어서 해요
  def insertWord(self):
    eng=input("단어를 입력하세요: ")
    kor=input("뜻을 입력하세요: ")
    lev=input("레벨을 입력하세요: ")
    word=Word(eng, kor, lev)
    wordList.append(word)
    print("등록되었습니다")

  def printAll(self):
    for data in wordList: #data에 워드 객체
      print(f"{data.getEng()}:{data.getKor()} 레벨 ({data.getLev()})")

  def saveAll(self):
    with open(filename,'w') as f:
      for data in wordList: #리스트에 들어가 있는 개수만큼
        f.write(f"{data.getEng()}/{data.getKor()}/{data.getLev()}\n")
      print("파일에 저장했습니다")

  def loadData(self):
    with open(filename,'r') as f:
      while True:
        line=f.readline()
        #apple/사과/1
        if line:
          #data=['apple', '사과', '1']
          data=line.strip().split("/")
          wordList.append(Word(data[0], data[1], data[2]))
        #데이터가 없을 때
        else:
          break
      print("데이터를 불러왔습니다")

  #과제 1 수정하기
  def updateWord(self):
    findword=input("수정할 단어를 입력하세요: ")
    for data in wordList:
      if data.getEng()==findword:
        data.setKor(input("뜻을 입력하세요: "))
        data.setLev(input("레벨을 입력하세요: "))
        print("수정되었습니다")
        break
      else:
        print("수정할 단어가 없습니다")

  #과제 2 삭제하기
  def deleteWord(self):
    findword=input("삭제할 단어를 입력하세요: ")
    for data in wordList:
      if data.getEng()==findword:
        wordList.remove(data)
        print("삭제되었습니다")
        break
      else:
        print("삭제할 단어가 없습니다")
`````

`````python
class Menu:
  def __init__(self):
    self.wordservice=WordService()

  def run(self):
    while True:
      try:
        menu = int(input("단어장 메뉴 1. 등록하기 2. 출력하기 3. 저장하기 4. 불러오기 5. 종료하기 6. 수정하기 7. 삭제하기: "))
        if menu == 1:
          self.wordservice.insertWord()
        elif menu==2:
          self.wordservice.printAll()
          pass
        elif menu==3:
          self.wordservice.saveAll()
        elif menu==4:
          self.wordservice.loadData()
        elif menu==5:
          print("프로그램을 종료합니다")
          break
        elif menu==6:
          self.wordservice.updateWord()
        elif menu==7:
          self.wordservice.deleteWord()
        else: #숫자 중에 엉뚱한 것을 사용
          print("다시 입력해 주세요")
      except Exception as e:
        print(e)
        print("다시 입력하세요")
`````

`````python
start=Menu()
start.run()
`````
