#**4. 함수**

함수의 매개변수와 반환 타입에 대한 어노테이션은 다음과 같이 합니다

```python
def add(a,b):
  return a+b
```

```python
print(add(10, 3))
print(add("십","삼"))
```

```python
def add(a:int,b:int) -> int:
  return a+b
```

```python
print(add(10, 3))
```

```python
print(add("십","삼")) #강제성이 없으므로 출력은 똑같이 된다
```

```python
add.__annotations__ #타입 어노테이션을 확인
```

#**5. 클래스**

```python
class Student:
  def __init__(self, name:str, age:int):
    self.name=name
    self.age=age
  def __str__(self) -> str:
    return "이름: " + self.name + ", 나이: "+ str(self.age)

def print_student(s:Student) -> None: #return None
  print(f"이름: {s.name}, 나이: {s.age} 🥞")
```

```python
student=Student("김사과",20)
print(student)
```

```python
print_student(student)
```

#**6. 옵셔널**

변수가 특정 타입이거나 None일 수 있는 경우 Optional을 사용합니다.

```python
from typing import Optional
```

```python
def find_score(name:str, scores:Dict[str, float]) -> Optional[float]: #평소에 float나오지만 None이 나올 수도 있다
  return scores.get(name)
```

```python
print(find_score("김사과",{'김사과':80.5,'오렌지':40.6,"반하나":90.5}))
```

```python
print(find_score("이메론",{'김사과':80.5,'오렌지':40.6,"반하나":90.5}))
```

#**7. 유니온**

여러 개의 타입이 허용될 수 있는 상황에 사용

```python
from typing import Union
```

```python
def toString(num:Union[int, float]) -> str:
  return str(num) #정수나 실수가 들어오면 문자열로 바꿔주는 함수
```

```python
print(toString(10))
print(toString(10.5))
```

```python
def toString(num:int|float) -> str: #파이썬 3.10 이후 #Union대신 or(|)을 쓴다
  return str(num) #정수나 실수가 들어오면 문자열로 바꿔주는 함수
```

```python
li: List[Union[int, str, float]]=[10,3.14,'안녕']
print(li)
```
