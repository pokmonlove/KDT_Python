##**1. 예외(Exception)**
프로그램 실행 중 발생할 수 있는 예상치 못한 문제 또는 오류 상황을 의미합니다. 예외가 발생하면 프로그램은 중단되기 때문에 이를 적절하게 처리하여 중단을 방지하거나 오류에 대한 정보를 사용자에게 제공해야 합니다.

````python
print(10/3)
print(5/0) ##error--> : division by zero
print(4/2)
````

#**2. 예외처리 기본 구조**

```
try:
    # 예외가 발생할 가능성이 있는 코드
    
except ExceptionType1:  # 'ExceptionType1'에는 실제 예외 유형이 들어갑니다.
    # ExceptionType1 예외가 발생했을 때 실행될 코드
    
except ExceptionType2:  # 'ExceptionType2'에는 다른 예외 유형이 들어갑니다.
    # ExceptionType2 예외가 발생했을 때 실행될 코드
    
# 추가적인 except 블록을 계속 추가할 수 있습니다.

else:
    # try 블록에서 예외가 발생하지 않았을 때 실행될 코드
    
finally:
    # 예외 발생 여부와 관계없이 항상 실행될 코드
```

````python
try:
  print(10/3)
  print(5/0)
  print(4/2)
except ZeroDivisionError:
  print("예외가 발생했습니다")
print("프로그램을 종료합니다") #비정상 종료 된 것 아님
````

````python
try:
  print(10/3)
  print(5/0)
  print(4/2)
except: #except 이름을 쓰지 않아도 되긴 한다.
  print("예외가 발생했습니다")
print("프로그램을 종료합니다") #비정상 종료 된 것 아님
````

````python
data = [10, 20, 30, 40, 50]
try:
  print(data[1])
  print(data[2])
  print(data[3])
  print(data[4])
  print(data[5])
except:
  print("인덱싱 지정이 올바르지 않습니다")
````

````python
#선생님 답안
try:
  data = [10, 20, 30, 40, 50]
  print(data[0])
  print(data[5])
  print(data[1]) #이 부분은 실행되지 않는다
except IndexError:
  print("인덱싱 지정이 올바르지 않습니다")
print("프로그램을 종료합니디")
````

````python
try:
  data = [10, 20, 30, 40, 50]
  #print(data[5])
  #print(int('안녕'))
  print(5/0)

except IndexError:
  print("인덱스 지정이 올바르지 않습니다")
except ValueError:
  print("값이 잘못 입력되었습니다")
except ZeroDivisionError:
  print("0으로 나눌 수 없습니다")
except:
  print("예외가 발생했어요") #except 전체는 맨밑에다가 넣는다
print("프로그램을 종료합니다")
````

````python
try:
  data = [10, 20, 30, 40, 50]
  print(data[5])
  #print(int('안녕'))
  #print(5/0)
  print(data[0])

except IndexError as e:
  print(e)
except ValueError as e:
  print(e)
except ZeroDivisionError as e:
  print(e)
except Exception as e:
  print(e) #except 전체는 맨밑에다가 넣는다
else: #에러발생 x여야 실행되는 부분
  print("에러가 발생하지 않은 정상적 프로그램")
finally:
  print("에러에 관계 없이 무조건 실행되는 문장")
print("프로그램을 종료합니다")
````

#**3. Exception 클래스**
Exception 클래스는 파이썬의 내장 예외 계층 구조에서 거의 모든 내장 예외의 기본 클래스입니다. 이 클래스는 사용자 정의 예외를 만들거나 특정 예외 유형을 잡기 위한 기본적인 인터페이스를 제공합니다.

예외사항을 직접 만들 수 있다.



1. 상속: 예외 유형은 Exception을 상속받아서 정의됩니다. 예를 들면 ValueError, TypeError, FileNotFoundError 등이 있습니다. 이 상속 구조 덕분에 except Exception 블록은 Exception을 상속받은 모든 예외를 캡처할 수 있습니다.

2. 메시지: 예외가 생성될 때, 일반적으로 오류 메시지를 함께 전달할 수 있습니다. 이 메시지는 예외 객체의 args 속성을 통해 접근 가능하며, 예외를 문자열로 변환할 때(예: str(e)) 해당 메시지가 반환됩니다.

````python
#raise : 예외 상황을 발생시킴
try:
  raise Exception("예외 발생")

except Exception as e:
  print(e)
````

````python
def func1():
  n=int(input("짝수를 입력하세요: "))
  if n%2==1:
    raise Exception("홀수를 입력했다") #문제 발생, 프로그램 강제 종료
  print(n)
````

````python
func1()
````

````python
try:
  func1()
except Exception as e:
  print("예외가 발생 ", e)
````

###문제2
* 사용자 정의 예외 클래스를 직접 만들고 활용해 보자
* 아래 클래스와 함수의 내용을 채우기
* 단, 나이의 범위는 아래와 같다
  * MZ세대는: 14살 ~ 43살
  * X세대: 44살 ~
  * 알파 세대: ~ 13살

````python
class AgeLimitError(Exception):
  def __init__(self, age, message="MZ세대 나이의 범위가 아님"):
    self.age=age
    self.message=message
    super().__init__(self.message) #부모에게 메시지 보내기
````

````python
def check_age(age):
  if age>43: raise AgeLimitError(age,"x세대")
  elif age<=13: raise AgeLimitError(age,"알파세대")
  else: return True
````

````python
ages=[17,60, 46, 20, 80, 55, 26,12,5]
for age in ages:
    try:
      check_age(age)
    except Exception as e:
      print(f"나이: {age}, 예외 발생", e)
    else:
      print("정상 출력", age, "살")
````

````python
#출력 결과
'''
17 나이 : 나이 범위 안에 포함
60 나이 에러: MZ 나이 범위보다 큼
...
5 나이 에러: MZ 나이 범위보다 작음
'''
````

````python
#선생님 답안
````

````python
class AgeLimitError(Exception):
  def __init__(self, age, message="MZ세대 나이의 범위가 아님"):
    self.age=age
    self.message=message
    super().__init__(self.message) #부모에게 메시지 보내기
````

````python
def check_age(age):
  if age<14: raise AgeLimitError(age,"MZ 나이 범위보다 작음")
  elif age>43: raise AgeLimitError(age,"MZ 나이 범위보다 큼")
  else: return f'나이: {age}, MZ 나이 범위 안에 포함'
````

````python
ages=[17,60, 46, 20, 80, 55, 26,12,5]
for age in ages:
    try:
      print(check_age(age))
    except AgeLimitError as e:
      print(f"{e.age} 나이 에러: {e}")
````
